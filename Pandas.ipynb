```python
nd=data.loc[(data['Type 1'] == 'Grass') & (data['Type 2'] == 'Poison') & (data['HP'] > 70)]
nd.to_csv('filtered_data.csv')
```

```python
#reset index and drop orignial index column
nd = nd.reset_index(drop = True)
nd
#drop "mega" in the names
nd = nd.loc[~nd['Name'].str.contains('Mega')]
nd
```

## Conditional Changes

```python
#change all 'Type 1' that are 'Fire' to 'Flamer'
data.loc[data['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
data
```

```python
nd= pd.read_csv('nd.csv')
nd
```

## Groupby for Aggregate Statistics

```python
#group data by 'Type 1' and in desecending order by 'Defense'
nd.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
# .sum and .count are some others
```
